Question 1

Convert 1D Array Into 2D Array
You are given a
0-indexed
1-dimensional (1D) integer array original, and two integers, m and n. You are tasked with creating a 2-dimensional (2D) array with m rows and n columns using
all
the elements from original.
The elements from indices 0 to n - 1 (
inclusive
) of original should form the first row of the constructed 2D array, the elements from indices n to 2 * n - 1 (
inclusive
) should form the second row of the constructed 2D array, and so on.
Return
an
m x n
2D array constructed according to the above procedure, or an empty 2D array if it is impossible

Picture1.png

Example 1:
Input:
original = [1,2,3,4], m = 2, n = 2
Output:
[[1,2],[3,4]]
Explanation:
The constructed 2D array should contain 2 rows and 2 columns.
The first group of n=2 elements in original, [1,2], becomes the first row in the constructed 2D array.
The second group of n=2 elements in original, [3,4], becomes the second row in the constructed 2D array.

```python
def convert_1d_array_into_2d_array(original, m, n):
  

  if len(original) < m * n:
    return []

  array = []
  for i in range(m):
    row = []
    for j in range(n):
      row.append(original[i * n + j])
    array.append(row)

  return array
```

Question 2

You have n coins and you want to build a staircase with these coins. The staircase consists of k rows where the ith row has exactly i coins. The last row of the staircase
may be
incomplete.
Given the integer n, return
the number of
complete rows
of the staircase you will build
.

Picture2.png

Example 1:

Input:
n = 5
Output:
2
Explanation:
Because the 3rd row is incomplete, we return 2.

```python
def arrange_coins(n):
 

  # Start with the first row, which has 1 coin.
  row = 1

  # Iterate over the remaining coins.
  for i in range(1, n):

    # Add the next coin to the current row.
    row += 1

    # If the current row is now too long, start a new row.
    if row > n:
      row = 1

  # Return the number of complete rows.
  return row
```

```python
>>> arrange_coins(5)
2
>>> arrange_coins(8)
3
```

 # Question 3

Given an integer array nums sorted in
non-decreasing
order, return
an array of
the squares of each number
sorted in non-decreasing order


Example 1:
Input:
nums = [-4,-1,0,3,10]
Output:
[0,1,9,16,100]
Explanation:
After squaring, the array becomes [16,1,0,9,100].
After sorting, it becomes [0,1,9,16,100].

```python
def sorted_squares(nums):
  

  # Create a new array to store the squares.
  squares = []

  # Iterate over the input array.
  for num in nums:
    # Square the current number and add it to the new array.
    squares.append(num * num)

  # Sort the new array.
  squares.sort()

  # Return the sorted array.
  return squares
```

```python
nums = [-4, -1, 0, 3, 10]

# Get the sorted squares of nums.
sorted_squares = sorted_squares(nums)

# Print the sorted squares.
print(sorted_squares)
```

 # Question 4

Given two
0-indexed
integer arrays nums1 and nums2, return
a list
answer
of size
2
where:

answer[0]
is a list of all
distinct
integers in
nums1
which are
not
present in
nums2
.

answer[1]
is a list of all
distinct
integers in
nums2
which are
not
present in
nums1.
Note
that the integers in the lists may be returned in
any
order.

Example 1:
Input:
nums1 = [1,2,3], nums2 = [2,4,6]
Output:
[[1,3],[4,6]]
Explanation:
For nums1, nums1[1] = 2 is present at index 0 of nums2, whereas nums1[0] = 1 and nums1[2] = 3 are not present in nums2. Therefore, answer[0] = [1,3].
For nums2, nums2[0] = 2 is present at index 1 of nums1, whereas nums2[1] = 4 and nums2[2] = 6 are not present in nums2. Therefore, answer[1] = [4,6].

```python
def find_distinct_numbers(nums1, nums2):
  

  # Create a set of all the numbers in nums1.
  nums1_set = set(nums1)

  # Create a set of all the numbers in nums2.
  nums2_set = set(nums2)

  # Find the intersection of the two sets.
  intersection_set = nums1_set.intersection(nums2_set)

  # Remove the intersection set from nums1_set.
  nums1_set = nums1_set - intersection_set

  # Remove the intersection set from nums2_set.
  nums2_set = nums2_set - intersection_set

  # Return a list of two lists, where the first list contains the elements of
  # nums1_set, and the second list contains the elements of nums2_set.
  return [list(nums1_set), list(nums2_set)]
```

```python
nums1 = [1, 2, 3]
nums2 = [2, 4, 6]

answer = find_distinct_numbers(nums1, nums2)

print(answer)
```

 # Question 5

Given two integer arrays arr1 and arr2, and the integer d,
return the distance value between the two arrays
.
The distance value is defined as the number of elements arr1[i] such that there is not any element arr2[j] where |arr1[i]-arr2[j]| <= d.
Example 1:
Input:
arr1 = [4,5,8], arr2 = [10,9,1,8], d = 2

Output:
2
Explanation:
For arr1[0]=4 we have:
|4-10|=6 > d=2
|4-9|=5 > d=2
|4-1|=3 > d=2
|4-8|=4 > d=2
For arr1[1]=5 we have:
|5-10|=5 > d=2
|5-9|=4 > d=2
|5-1|=4 > d=2
|5-8|=3 > d=2
For arr1[2]=8 we have:
|8-10|=2 <= d=2
|8-9|=1 <= d=2
|8-1|=7 > d=2
|8-8|=0 <= d=2

```python
def find_distance_value(arr1, arr2, d):
  

  count = 0
  for i in range(len(arr1)):
    found_match = False
    for j in range(len(arr2)):
      if abs(arr1[i] - arr2[j]) <= d:
        found_match = True
        break

    if not found_match:
      count += 1

  return count
```

```python
arr1 = [4, 5, 8]
arr2 = [10, 9, 1, 8]
d = 2

print(find_distance_value(arr1, arr2, d))
```

 # Question 6

Given an integer array nums of length n where all the integers of nums are in the range [1, n] and each integer appears
once
or
twice
, return
an array of all the integers that appears
twice
.
You must write an algorithm that runs in O(n) time and uses only constant extra space.

Example 1:
Input:
nums = [4,3,2,7,8,2,3,1]
Output:
[2,3]

```python
def find_duplicates(nums):
  

  # Create a hash table to store the number of occurrences of each element.
  seen = {}

  # Iterate over the array and increment the count of each element in the hash table.
  for num in nums:
    if num in seen:
      seen[num] += 1
    else:
      seen[num] = 1

  # Find all elements in the hash table with a count of greater than 1.
  duplicates = []
  for num, count in seen.items():
    if count > 1:
      duplicates.append(num)

  return duplicates
```

```python
nums = [4, 3, 2, 7, 8, 2, 3, 1]

print(find_duplicates(nums))
```

 # Question 7

Suppose an array of length n sorted in ascending order is
rotated
between 1 and n times. For example, the array nums = [0,1,2,4,5,6,7] might become:

[4,5,6,7,0,1,2] if it was rotated 4 times.

[0,1,2,4,5,6,7] if it was rotated 7 times.
Notice that
rotating
an array [a[0], a[1], a[2], ..., a[n-1]] 1 time results in the array [a[n-1], a[0], a[1], a[2], ..., a[n-2]].
Given the sorted rotated array nums of
unique
elements, return
the minimum element of this array
.

You must write an algorithm that runs in O(log n) time.
Example 1:
Input:
nums = [3,4,5,1,2]
Output:
1
Explanation:
The original array was [1,2,3,4,5] rotated 3 times

```python
def find_minimum_in_rotated_sorted_array(nums):
  

  # Find the index of the smallest element in the array.
  start = 0
  end = len(nums) - 1
  while start < end:
    mid = (start + end) // 2
    if nums[mid] > nums[end]:
      start = mid + 1
    else:
      end = mid

  # Return the smallest element.
  return nums[start]
```

```python
nums = [3, 4, 5, 1, 2]

print(find_minimum_in_rotated_sorted_array(nums))
```

# Question 8

An integer array original is transformed into a
doubled
array changed by appending
twice the value
of every element in original, and then randomly
shuffling
the resulting array.
Given an array changed, return original
if
changed
is a
doubled
array. If
changed
is not a
doubled
array, return an empty array. The elements in
original
may be returned in
any
order
.
Example 1:
Input:
changed = [1,3,4,2,6,8]
Output:
[1,3,4]

```python
def find_original_array(changed):
  

  # Create a map to store the frequencies of the elements in the doubled array.
  frequencies = {}
  for element in changed:
    frequencies[element] = frequencies.get(element, 0) + 1

  # Find the smallest element in the doubled array.
  smallest_element = min(changed)

  # Create a list to store the elements of the original array.
  original = []

  # Iterate over the doubled array.
  for element in changed:
    # If the element is the smallest element, add it to the original array.
    if element == smallest_element:
      original.append(element)

    # If the element is twice the value of an element in the original array,
    # remove that element from the original array.
    elif element == 2 * smallest_element:
      frequencies[smallest_element] -= 1

    # If the element is not in the original array, return an empty array.
    elif frequencies[element] == 0:
      return []

  # Return the original array.
  return original
```

```python
>>> find_original_array([1, 3, 4, 2, 6, 8])
[1, 3, 4]
>>> find_original_array([1, 2, 4, 8])
[]
>>> find_original_array([1])
[]
```
